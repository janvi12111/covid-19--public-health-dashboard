# COVID-19 Public Health Analysis
Companion notebook to `python/clean_data.py` and `python/analysis.py`.
Run the two scripts once to populate `data/processed/` and `python/images/`, then explore interactively below.

```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
sns.set_theme(style='whitegrid')
daily = pd.read_csv('../data/processed/daily_stats.csv', parse_dates=['date'])
countries = pd.read_csv('../data/processed/countries.csv')
daily.head()
```

## Global daily new cases

```python
g = daily.groupby('date')['new_cases'].sum()
g.rolling(7).mean().plot(figsize=(12,5), title='Global new cases (7-day MA)');
```

## Per-country peak day

```python
peaks = daily.loc[daily.groupby('country_name')['new_cases_ma7'].idxmax(),
                  ['country_name','date','new_cases_ma7']]
peaks.sort_values('new_cases_ma7', ascending=False).head(10)
```

## Case-fatality rate ranking

```python
cfr = (daily.groupby('country_name')
            .agg(total_cases=('total_cases','max'), total_deaths=('total_deaths','max')))
cfr['cfr_pct'] = cfr['total_deaths'] / cfr['total_cases'] * 100
cfr.sort_values('cfr_pct', ascending=False).head(10).round(2)
```
